```python
data = df.sample(n=50, random_state=42)
```

```python
print(data.head())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Quantity'], bins=30, kde=True)
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Movement Type', y='Quantity', data=data)
plt.title('Quantity by Movement Type')
plt.xlabel('Movement Type')
plt.ylabel('Quantity')
plt.show()
```

```python
material_qty = data.groupby('Material')['Quantity'].sum().reset_index().sort_values(by='Quantity', ascending=False)

plt.figure(figsize=(14, 8))
sns.barplot(x='Quantity', y='Material', data=material_qty.head(20))
plt.title('Top 20 Materials by Total Quantity')
plt.xlabel('Total Quantity')
plt.ylabel('Material')
plt.show()
```

```python
numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()

# Compute the correlation matrix
correlation_matrix = data[numeric_cols].corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
vendor_qty = data.groupby('Vendor Code')['Quantity'].sum().reset_index().sort_values(by='Quantity', ascending=False)

plt.figure(figsize=(14, 8))
sns.barplot(x='Quantity', y='Vendor Code', data=vendor_qty.head(20))
plt.title('Top 20 Vendors by Total Quantity')
plt.xlabel('Total Quantity')
plt.ylabel('Vendor Code')
plt.show()
```

```python
data['Posting Month'] = data['Posting Date']

monthly_qty = data.groupby('Posting Month')['Quantity'].sum().reset_index()

plt.figure(figsize=(14, 8))
sns.lineplot(x='Posting Month', y='Quantity', data=monthly_qty)
plt.title('Trend of Quantity over Time')
plt.xlabel('Posting Month')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45)
plt.show()
```
